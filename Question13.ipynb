```python
import pandas as pd
import matplotlib.pyplot as plt

# Load and prepare data
file_path = "question 13 revenue_drop_analysis.xlsx"
df = pd.read_excel(file_path)
df["Date"] = pd.to_datetime(df["Date"])
df.set_index("Date", inplace=True)

# 1. Area Chart: Monthly Product/Service Volume
plt.figure(figsize=(10, 5))
df["Product/Service Volume"].resample("M").sum().plot(kind="area", color="skyblue", title="Monthly Product/Service Volume")
plt.xlabel("Month")
plt.ylabel("Volume")
plt.tight_layout()
plt.show()

# 2. Line Chart: Monthly Churn Rate
plt.figure(figsize=(10, 5))
df["Churn Rate"].resample("M").mean().plot(marker="o", color="purple", title="Monthly Churn Rate Trend")
plt.xlabel("Month")
plt.ylabel("Churn Rate (%)")
plt.tight_layout()
plt.show()



```
